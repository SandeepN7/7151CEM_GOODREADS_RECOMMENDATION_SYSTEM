```python
from selenium import webdriver
import time
import base64
import requests
from bs4 import BeautifulSoup
import csv
from csv import writer
import json
from selenium.webdriver.common.by import By
```

```python
browser = webdriver.Chrome()
browser.maximize_window()
ab = "https://www.goodreads.com/list/show/1.Best_Books_Ever"
u = "https://www.goodreads.com"
browser.get(ab)
with open("Goodreads_URL.txt", 'a') as file:        
    for i in range(1,101):

        browser.get("https://www.goodreads.com/list/show/1.Best_Books_Ever?page=%d"%(i))
        a = browser.page_source
        resp = BeautifulSoup(a,"html.parser")
        s = resp.find_all(class_ = 'bookTitle')
        for s1 in s:
            href = s1['href']
            url = u + href
            print(url, file=file)
        try:
            browser.find_element(By.XPATH, "/html/body/div[3]/div/div/div[1]/button/img").click()
        except:
            pass        
```
