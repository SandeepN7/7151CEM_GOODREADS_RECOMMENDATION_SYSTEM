```python
from selenium import webdriver
import time
import base64
import requests
from bs4 import BeautifulSoup
import csv
from csv import writer
import json
from selenium.webdriver.common.by import By
```

```python
browser = webdriver.Chrome()
browser.maximize_window()
with open("GoodReads_book_details.csv","a", newline="") as outfile:
    csv_writer = writer(outfile)
    csv_writer.writerow(["Book_ID","Title","Authors","Total_Pages","No. of Reviews","Average_Rating","No. of Ratings","ISBN","ISBN13","Language","Genres"])
    
    f = open('Goodreads_URL.txt')
    for line in f:
        browser.get(line)
        try:
            browser.find_element(By.XPATH, "/html/body/div[3]/div/div/div[1]/button/img").click()
        except:
            pass

        browser.find_element(By.XPATH, "/html/body/div[2]/div[3]/div[1]/div[2]/div[4]/div[1]/div[2]/div[5]/div[3]/a[1]").click()
        time.sleep(1)
        a = browser.page_source
        resp=BeautifulSoup(a,"html.parser")

        try:
            t = resp.find('cite')
            title = t.text
#             ti = title.encode("utf-8")
        except:
            ti = ""
        
    #     print("Title:", title)
        
        try:
            auth = resp.find(class_='authorName__container')
            auth1 = auth.find(class_='authorName')
            author = auth1.text
        except:
            author = ""
        
        
    #     print("Author:", auth1.text)

        try:
            r = resp.find(itemprop = 'ratingValue')
            r1 = r.text
            rating = r1.strip()
        except:
            rating = ""
        
    #     print("Ratings:", rating)
        
        try:
            rd = resp.find(itemprop = 'ratingCount')
            rd1 = rd['content']
        except:
            rd1 = ""
        
    #     print("Total_Ratings:", rd1)

        try:
            rc = resp.find(itemprop = 'reviewCount')
            rc1 = rc['content']
        except:
            rc1 = ""
        
    #     print("Reviews:", rc1)

        try:
            pages = resp.find(property = 'books:page_count')
            p1 = pages['content'] 
        except:
            p1 = ""
          
    #     print("Total Pages:", p1) 

        i = resp.find(class_ = 'asyncPreviewButtonContainer')
        b_id = i['data-book-id']
        
    #     print("Book ID:", b_id)        

        i1 = resp.find(itemprop = 'isbn')
        if (i1 != None):
            isbn13 = i1.text
        else:
            isbn13 = ""
#         print("ISBN13:", isbn13)

        try: 
            genre = resp.find_all(class_ = 'elementList')
            ar = []
            for g in genre:
                g1 = g.find(class_ = 'left')            
                if(g1 == None):
                    continue
                g2 = g1.find_all('a')
                for gen in g2:
                    ar.append(gen.text)            

        except:
            pass

        details = resp.find_all(class_='clearFloats')
        isbn_final = ""
        try:
        
            for d in details:       
                s = d.find(class_='infoBoxRowTitle')
                s1 = s.text            

                if(s1 == 'Edition Language'):
                    lang = d.find(class_ ='infoBoxRowItem')
                    lan = lang.text
                
                if(s1 == 'ISBN'):
                    isb = d.find(class_ ='infoBoxRowItem')
                    isbn = isb.text
                    isbn_temp = isbn.strip()                    
                    isbn_final = isbn_temp[0:10]
                    
                else:
                    continue
        except:
            pass
        # print(title)
        csv_writer.writerow([b_id,title,author,p1,rc1,rating,rd1,isbn_final,isbn13,lan,ar])
```
